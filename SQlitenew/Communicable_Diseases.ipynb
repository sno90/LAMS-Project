```python
# Python SQL toolkit and Object Relational Mapper
import pandas as pd
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
```

```python
engine = create_engine("sqlite:///Resources/Communicable_Disease_VA.sqlite")
```

```python
df = pd.read_csv("Resources/Communicable_Disease_VA.csv").rename(columns = {"Rate per 100,000": "Rate"})
df
```

```python
df.to_sql("diseases",con = engine, if_exists = "append", index = False)
```

```python
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
conn = engine.connect()
```

```python
conn.execute("DROP TABLE diseases")
```

```python
conn.execute("""CREATE TABLE DISEASES(
            Year Integer not null, 
            State Varchar,
            Region Varchar,
            District Varchar,
            County Varchar not null,
            Condition Varchar not null,
            Population Varchar,
            Cases Integer,
            Rate Integer, 
            id serial Primary Key)""")
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```
